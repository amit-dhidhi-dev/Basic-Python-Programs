```python
# @title ‚ñ∂Ô∏è convert . WAV TO .MP3
import subprocess
import os

def wav_to_mp3(input_wav_path, output_mp3_path=None, bitrate="192k"):
    """
    Convert a WAV file to MP3 using FFmpeg.

    Args:
        input_wav_path (str): Path to input WAV file.
        output_mp3_path (str, optional): Path to save MP3 file.
            Defaults to same name as input with .mp3 extension.
        bitrate (str, optional): Bitrate for MP3 file. Default is "192k".

    Returns:
        str: Path to generated MP3 file.
    """
    # Check if input exists
    if not os.path.isfile(input_wav_path):
        raise FileNotFoundError(f"WAV file not found: {input_wav_path}")

    # Set default output path if not provided
    if output_mp3_path is None:
        base_name = os.path.splitext(input_wav_path)[0]
        output_mp3_path = f"{base_name}.mp3"

    # Run ffmpeg command
    cmd = [
        "ffmpeg",
        "-y",  # overwrite if exists
        "-i", input_wav_path,
        "-b:a", bitrate,
        output_mp3_path
    ]

    try:
        subprocess.run(cmd, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        print(f"Conversion successful: {output_mp3_path}")
    except subprocess.CalledProcessError as e:
        print("Error during conversion:")
        print(e.stderr.decode())
        return None

    return output_mp3_path
```

```python
# @title ‚ñ∂Ô∏è DOWNLOAD MP3
mp3_file = wav_to_mp3("output.wav", bitrate="256k")

# Download the MP3
from google.colab import files
files.download(mp3_file)
```
